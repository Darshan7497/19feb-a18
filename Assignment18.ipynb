# Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible Python web framework that allows developers to build web applications quickly and easily. It was created by Armin Ronacher. Advantages of using Flask Framework:

Lightweight and flexible: Flask is a lightweight framework that allows developers to create web applications quickly and easily without the need for extensive setup or configuration.

Easy to learn: Flask has a simple and intuitive interface that is easy to learn for developers with basic knowledge of Python programming.

Modular design: Flask has a modular design that allows developers to choose and use only the components they need. This makes Flask very flexible and adaptable to different project requirements.

Extensible: Flask can be easily extended with additional libraries and modules, making it easy to integrate with other technologies and frameworks.

Built-in development server: Flask has a built-in development server that allows developers to test and debug their applications easily without the need for additional setup.

Large community: Flask has a large and active community of developers who contribute to the framework by creating extensions, tutorials, and other resources.

Support for multiple databases: Flask supports multiple databases such as SQLite, MySQL, PostgreSQL, and MongoDB, which makes it easy to work with different data storage solutions.

# Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to functions that are responsible for generating the content that will be displayed when the URL is visited. Flask uses decorators to map a URL to a function.For example, consider the following Flask app:

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/about')
def about():
    return 'About page'

if __name__ == '__main__':
    app.run(host ="0.0.0.0")
```

In this example, we have defined two routes using the @app.route decorator. The first route maps the URL '/' to the index function, which returns the string 'Hello, World!' when the URL is visited. The second route maps the URL '/about' to the about function, which returns the string 'About page' when the URL is visited.

We use app routes in Flask to define the different pages and functionality of our web application. By mapping URLs to functions, we can create a hierarchical structure for our application, with different pages accessible via different URLs. This makes it easy to organize our code and create a logical flow for our users as they navigate our application.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL based on the name of a function and any arguments passed to it. The url_for() function takes the name of the function as its first argument and any keyword arguments as additional arguments.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the homepage!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/contact')
def contact():
    return 'Please contact us at contact@example.com'

@app.route('/user/<username>')
def user_profile(username):
    return f'This is the profile page for {username}'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))             # Output: /
        print(url_for('about'))            # Output: /about
        print(url_for('contact'))          # Output: /contact
        print(url_for('user_profile', username='darshan'))   # Output: /user/darshan
```
